```python
def remove_basis_states_from_superposition(state, array_of_bitstrings):

    mask = np.zeros(state.state_matrix.shape[0], dtype=bool)
    for basis_state in array_of_bitstrings:
       mask = np.logical_or(mask, np.sum(state.state_matrix == basis_state, axis=1) == state.state_matrix.shape[1])

    return sym.QuantumState(state.state_matrix[~mask], state.state_op.coeff_vec[~mask])
```

```python
def get_Qz_state(N):
    n_qubits = 2*N

    bitstrings = (
                    (
                            np.arange(2 ** n_qubits).reshape([-1, 1]) &
                            (1 << np.arange(n_qubits))[::-1]
                    ) > 0).astype(int)

    sign_array = np.array([(-1)**(i%2) for i in range(n_qubits)])
    Qz_mask = np.sum(bitstrings*sign_array, axis=1) == 0

    valid_bitstrings = bitstrings[Qz_mask, :]
    return valid_bitstrings

def get_valid_B(array_bitstrings, B_value):
    B_arr = (-1)**array_bitstrings * 1/4
    B_mask = np.sum(B_arr, axis=1) == B_value
    return array_bitstrings[B_mask, :]
```

```python
B_val = 1
valid_Qz = get_Qz_state(N)
valid_Qz_and_B = get_valid_B(valid_Qz, B_val)
valid_Qz_and_B
```

```python
single_solutions=[]
for bitstring in valid_Qz_and_B:
    if (Qy * sym.QuantumState(bitstring.reshape([1,-1]))).n_terms==0:
        single_solutions.append(bitstring)

single_solutions = np.array(single_solutions)
single_solutions
```

```python
# other illegal solutions
illegal_B_vals = np.setdiff1d([1,-1, 0], B_val)

remove_Bneg1 = get_valid_B(valid_Qz, illegal_B_vals[0])
remove_B1 = get_valid_B(valid_Qz, illegal_B_vals[1])


illegal = single_solutions
illegal = np.vstack((illegal, remove_Bneg1))
illegal = np.vstack((illegal, remove_B1))
illegal
```

```python
solutions = {}
for i in range(arr.shape[0]):
    #state =  sym.QuantumState(arr[i,:].reshape([1,-1]), [1]) 
    state =  sym.QuantumState(arr_updated[i,:].reshape([1,-1]), [1]) 
    
    # solutions[i] = Ry.dagger * state
    rotated_state = Ry.dagger * state
    rotated_state_no_ind_sol = remove_basis_states_from_superposition(rotated_state, 
                                                                      illegal)
    solutions[i] = rotated_state_no_ind_sol
solutions
```

```python
solutions_real = {}
for key, state in solutions.items():
    if state.n_terms>0:
        mask = np.abs(state.state_op.coeff_vec.real)>1e-12
        real_state_matrix = state.state_matrix[mask]
        coeff = state.state_op.coeff_vec[mask]
        solutions_real[i] = sym.QuantumState(real_state_matrix, coeff).normalize

# dictionary hashing removes duplicated terms!
solutions_real
```
